# Load the required packages

```python
import os
import scanpy as sc
import pandas as pd
import numpy as np
import RedeFISH
import sys
os.environ["CUDA_VISIBLE_DEVICES"]="0"
```

# Load the required datasets

```python
device = "cuda"

# The transcripts file must contain the x, y, and gene columns of the transcripts.
data = pd.read_csv("example_data/st_data.HumanLungCancer.csv")
# The Scanpy h5ad file of gene expression must contain .obs['annotation'] corresponding to the annotation of each cell.
sc_data = sc.read("example_data/sc_data.HumanLungCancer.h5ad")
# Polygonal regions of the nucleus boundary, containing cell_id, x, and y columns, with the same cell_id considered to be from the same nucleus boundary.
nuclei_boundary_prior = pd.read_csv("example_data/nuclei_boundary.HumanLungCancer.csv")

sc_data.var_names_make_unique()
coordinate = np.asarray([data['x'], data['y']]).T
```

# Running RedeFISH using nuclear boundary prior

```python
if __name__ == "__main__":

    # scale refer to prior cell radius, typically set to approximately 6.5 µm, which corresponds to about 30 pixels in this dataset.
    scale = 6.5

    # If the input prior is the nucleus region, set boundary_prior = True.
    model = RedeFISH.alignment(coordinate,
                               list(data['gene']),
                               nuclei_boundary_prior,
                               sc_data,
                               scale = scale,
                               boundary_prior = True,
                               worker_number = 16,
                               output_dir = "test_segmentation_with_nuclear_boundary_prior",
                               device=device).to(device)

    model.train()
```
